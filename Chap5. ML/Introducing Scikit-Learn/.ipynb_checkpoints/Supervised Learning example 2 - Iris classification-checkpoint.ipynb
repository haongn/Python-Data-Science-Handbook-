Given a model trained on a portion of the Iris data,
how well can we predict the remaining labels?

```python
# generative model: Gaussian naive Bayes
```

```python
# Gaussian NB has no hyperparameter to choose
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
type(iris)
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris, hue = 'species', size = 1.5)
```

```python
# extract features matrix and target array from dataframe
X_iris = iris.drop('species', axis = 1)  # loại bỏ theo cột
X_iris.shape
```

```python
y_iris = iris['species']
y_iris.shape
```

```python
# split the data into a training set and testing set, using utility function
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state = 1)
```

```python
# follow our recipe to predict the labels: 
from sklearn.naive_bayes import GaussianNB        # 1. choose model class
model = GaussianNB()                              # 2. instantiate model 
model.fit(Xtrain, ytrain)                         # 3. fit model to data
y_model = model.predict(Xtest)                    # 4. predict on new data
```

```python
# use utility function to see the fraction of predicted labels that match their true value
from sklearn.metrics import accuracy_score 
accuracy_score(ytest, y_model)     # xem tỉ lệ đánh nhãn đúng là bao nhiêu
```
